```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'df' contains your dataset with columns: 'area', 'bedrooms', 'age', 'price'

# Load the dataset
# df = pd.read_csv("homeprices.csv")  # Load your CSV file

# Ensure 'area' values are numeric (remove commas and convert to float)
df['area'] = df['area'].replace(',', '', regex=True).astype(float)

# Handle missing values in 'bedrooms' (if any)
median_bedrooms = df['bedrooms'].median()
df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)

# Prepare the features and target variable
X = df[['area', 'bedrooms', 'age']]  # Features (independent variables)
y = df['price']  # Target variable

# Initialize and train the linear regression model
regression_model = LinearRegression()
regression_model.fit(X, y)

# Make predictions
predicted_prices = regression_model.predict(X)

# Plotting the graph
plt.figure(figsize=(8, 6))

# Scatter plot for actual vs predicted prices
plt.scatter(y, predicted_prices, color='blue')
plt.title('Actual Prices vs Predicted Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'df' contains your dataset with columns: 'area', 'bedrooms', 'age', 'price'

# Load the dataset
# df = pd.read_csv("homeprices.csv")  # Load your CSV file

# Ensure 'area' values are numeric (remove commas and convert to float)
df['area'] = df['area'].replace(',', '', regex=True).astype(float)

# Handle missing values in 'bedrooms' (if any)
median_bedrooms = df['bedrooms'].median()
df['bedrooms'] = df['bedrooms'].fillna(median_bedrooms)

# Prepare the features and target variable
X = df[['area', 'bedrooms', 'age']]  # Features (independent variables)
y = df['price']  # Target variable

# Initialize and train the linear regression model
regression_model = LinearRegression()
regression_model.fit(X, y)

# Make predictions
predicted_prices = regression_model.predict(X)

# Plotting the graph
plt.figure(figsize=(8, 6))

# Scatter plot for actual vs predicted prices
plt.scatter(y, predicted_prices, color='blue', label='Actual vs Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')

# Adding a diagonal line (ideal case)
min_val = min(min(y), min(predicted_prices))
max_val = max(max(y), max(predicted_prices))
plt.plot([min_val, max_val], [min_val, max_val], color='red', label='Ideal line')

plt.legend()
plt.grid(True)
plt.show()
```
